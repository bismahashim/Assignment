```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book = pd.read_csv('book.csv',encoding='Latin1')
```

```python
book
```

```python
books=book.iloc[:,1:]
books
```

```python
books.sort_values(['User.ID'])
```

```python
len(books['User.ID'].unique())
```

```python
len(books['Book.Title'].unique())
```

```python
books1=books.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books1
```

```python
books1.index=books['User.ID'].unique()
books1
```

```python
books1.fillna(0,inplace=True)
books1
```

```python
user_sim=1-pairwise_distances(books1.values,metric='cosine')
user_sim
```

```python
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
user_sim2.index=books['User.ID'].unique()
user_sim2.columns=books['User.ID'].unique()
user_sim2
```

```python
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
user_sim2.idxmax(axis=1)
```
