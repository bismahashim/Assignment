```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
Wine= wine.iloc[:,1:]
Wine
```

```python
Wine_normal = scale(Wine)
Wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(Wine_normal)
pca_values
```

```python
pca.components_ 
```

```python
var = pca.explained_variance_ratio_
var
```

```python
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="red")
```

```python
final_df=pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
wine= pd.read_csv("wine.csv")
```

```python
wine.head()
```

```python
wine_normal = scale(Wine)
wine_normal
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_normal,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_normal),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine2=wine.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

```python
from sklearn.cluster import KMeans
```

```python

wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
clusters3=KMeans(3,random_state=30).fit(wine_normal)
clusters3
```

```python
clusters3.labels_
```

```python
wine3=wine.copy()
wine3['clusters3id']=clusters3.labels_
wine3
```
